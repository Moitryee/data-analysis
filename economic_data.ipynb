## Discription of this Data
This data is about consumer price index.Consumer Price Indices (CPI) measure changes over time in general level of prices of goods and services that households acquire for the purpose of consumption. this index is made after considering all the consuming products which are consumed by consumers not considering manufactured products(which is used by firm such as hardware of server used by company). this data taken from the year 2013 - 2023. there are total 10 years and 5 months. here we are considering base year 2010.



## Discription of the columns
1. Sector: this column has three catagory urban,rural,urban+rural (catagorical data)
2. year : from 2013 to 2023 (catagorical data)
3. month: from january to december , (catagorical data)
all other columns are given price level on the specific time period.

### 2.In which month in every year have the highest price in India?

```python
## finding th sum of values of each row
df['Sum'] = df.iloc[:,3:-1].sum(axis=1,)
df
```

```python
## groupby by year 
df_new=df[df["Sector"]=="Rural+Urban"].groupby("Year")["Sum"].max().reset_index()
```

```python
## to get the full rows we have to merge the data
merged_df = df.merge(df_new, on=["Year", "Sum"], how="inner")
```

```python
## result is here
merged_df
```

```python
#lets visualize it
sns.countplot(x="Month",data=merged_df)

plt.ylabel("NoOfMonths")
plt.title("countPlot")
```
